```python
def seasonal_time_process_all(hormone,sex,dtype='quantile'):

    file    = '../data/Clalit_seasonality_Nov19/clock.' + hormone +'.healthy_no_meds.' + dtype + '.with_infection.0.05.20_80.csv'
    df  = pd.read_csv(file) 
    df  = df.loc[df['type'] ==sex]
    
    df['hour'] = df['hour']+12
    df.loc[df['hour']>24,['hour']] = df.loc[df['hour']>24,['hour']] - 24
    df = df.loc[df['hour']>6]
    df = df.loc[df['hour']<19]
    df = df.loc[df['n']>30]
    
    seas_pooled_mean = []
    seas_pooled_std  = []
    seas_pooled_n    = []

    for seas in ['winter','spring','summer','fall']:
        mdf = df.loc[df.season==seas]
        seas_pooled_mean.append(np.sum(mdf['mean']*mdf['n'])/np.sum(mdf['n']))
        seas_pooled_std.append(np.sqrt(np.sum(mdf['std']**2*(mdf['n']-1))/(np.sum(mdf['n']) - len(mdf['n']))))
        seas_pooled_n.append(np.sum(mdf['n']))

    both_res = pd.DataFrame({'seasons': ['winter','spring','summer','fall'],
                             'mean' : seas_pooled_mean,
                             'std'  : seas_pooled_std,
                             'n'    : seas_pooled_n})
    return both_res


def seasonal_time_process_9am(hormone,sex,dtype='quantile'):
    file    = '../data/Clalit_seasonality_Nov19/clock.' + hormone +'.healthy_no_meds.' + dtype + '.with_infection.0.05.20_80.csv'
    df  = pd.read_csv(file)  
    df  = df.loc[df['type'] ==sex]
    
    df['hour'] = df['hour']+12
    df.loc[df['hour']>24,['hour']] = df.loc[df['hour']>24,['hour']] - 24
    df = df.loc[df['n']>30]
    
    df  = df.loc[df['hour'] == 9]
    seas_pooled_mean = []
    seas_pooled_std  = []
    seas_pooled_n    = []
    for seas in ['winter','spring','summer','fall']:
        mdf = df.loc[df.season==seas]
        seas_pooled_mean.append(np.sum(mdf['mean']*mdf['n'])/np.sum(mdf['n']))
        seas_pooled_std.append(np.sqrt(np.sum(mdf['std']**2*(mdf['n']-1))/(np.sum(mdf['n']) - len(mdf['n']))))
        seas_pooled_n.append(np.sum(mdf['n']))

    both_res = pd.DataFrame({'seasons': ['winter','spring','summer','fall'],
                             'mean' : seas_pooled_mean,
                             'std'  : seas_pooled_std,
                             'n'    : seas_pooled_n})
    return both_res

def seasonal_time_process_8am(hormone,sex,dtype='quantile'):
    file    = '../data/Clalit_seasonality_Nov19/clock.' + hormone +'.healthy_no_meds.' + dtype + '.with_infection.0.05.20_80.csv'
    df  = pd.read_csv(file)  
    df  = df.loc[df['type'] ==sex]
    
    df['hour'] = df['hour']+12
    df.loc[df['hour']>24,['hour']] = df.loc[df['hour']>24,['hour']] - 24
    df = df.loc[df['n']>30]
    
    df  = df.loc[df['hour'] == 8]
    seas_pooled_mean = []
    seas_pooled_std  = []
    seas_pooled_n    = []
    for seas in ['winter','spring','summer','fall']:
        mdf = df.loc[df.season==seas]
        seas_pooled_mean.append(np.sum(mdf['mean']*mdf['n'])/np.sum(mdf['n']))
        seas_pooled_std.append(np.sqrt(np.sum(mdf['std']**2*(mdf['n']-1))/(np.sum(mdf['n']) - len(mdf['n']))))
        seas_pooled_n.append(np.sum(mdf['n']))

    both_res = pd.DataFrame({'seasons': ['winter','spring','summer','fall'],
                             'mean' : seas_pooled_mean,
                             'std'  : seas_pooled_std,
                             'n'    : seas_pooled_n})
    return both_res

def seasonal_time_process_sunrise_phase(hormone,sex,dtype='quantile'):
    ## winter sunrise at 6:35       - use 9AM
    ## spring sunrise at 5:40/6:30  - use 8AM,9AM
    ## summer sunrise at 5:34       - use 8AM
    ## fall sunrise at 6:26/5:50    - use 8AM,9AM
    
    file    = '../data/Clalit_seasonality_Nov19/clock.' + hormone +'.healthy_no_meds.' + dtype + '.with_infection.0.05.20_80.csv'
    df  = pd.read_csv(file)  
    df  = df.loc[df['type'] ==sex]
    
    df['hour'] = df['hour']+12
    df.loc[df['hour']>24,['hour']] = df.loc[df['hour']>24,['hour']] - 24
    df = df.loc[df['hour']>6]
    df = df.loc[df['hour']<11]
    df = df.loc[df['n']>30]
    
    seas_pooled_mean = []
    seas_pooled_std  = []
    seas_pooled_n    = []
    for seas in ['winter','spring','summer','fall']:
        if seas == 'winter':
            hdf  = df.loc[df['hour'] == 9].copy()
        if seas == 'spring':
            hdf  = df.loc[((df['hour'].values==8)*1 + (df['hour'].values==9)*1)==1].copy()  
        if seas == 'summer':
            hdf  = df.loc[df['hour'] == 8].copy()    
        if seas == 'fall':
            hdf  = df.loc[((df['hour'].values==8)*1 + (df['hour'].values==9)*1)==1].copy()   
        mdf = hdf.loc[hdf.season==seas]
        seas_pooled_mean.append(np.sum(mdf['mean']*mdf['n'])/np.sum(mdf['n']))
        seas_pooled_std.append(np.sqrt(np.sum(mdf['std']**2*(mdf['n']-1))/(np.sum(mdf['n']) - len(mdf['n']))))
        seas_pooled_n.append(np.sum(mdf['n']))

    both_res = pd.DataFrame({'seasons': ['winter','spring','summer','fall'],
                             'mean' : seas_pooled_mean,
                             'std'  : seas_pooled_std,
                             'n'    : seas_pooled_n})
    return both_res

def seasonal_time_process_minimum(hormone,sex,dtype='quantile'):

    file    = '../data/Clalit_seasonality_Nov19/clock.' + hormone +'.healthy_no_meds.' + dtype + '.with_infection.0.05.20_80.csv'
    df  = pd.read_csv(file) 
    df  = df.loc[df['type'] ==sex]
    
    df['hour'] = df['hour']+12
    df.loc[df['hour']>24,['hour']] = df.loc[df['hour']>24,['hour']] - 24
    df = df.loc[df['hour']>6]
    df = df.loc[df['hour']<11]
    df = df.loc[df['n']>30]
    
    seas_pooled_mean = []
    seas_pooled_std  = []
    seas_pooled_n    = []

    for seas in ['winter','spring','summer','fall']:
        mdf = df.loc[df.season==seas].copy()
        il  = np.abs(np.diff(mdf['mean'].values)).argmin()
        mdf = mdf.iloc[[il,il+1]]
        seas_pooled_mean.append(np.sum(mdf['mean']*mdf['n'])/np.sum(mdf['n']))
        seas_pooled_std.append(np.sqrt(np.sum(mdf['std']**2*(mdf['n']-1))/(np.sum(mdf['n']) - len(mdf['n']))))
        seas_pooled_n.append(np.sum(mdf['n']))

    both_res = pd.DataFrame({'seasons': ['winter','spring','summer','fall'],
                             'mean' : seas_pooled_mean,
                             'std'  : seas_pooled_std,
                             'n'    : seas_pooled_n})
    return both_res

def seasonal_time_process_equal_time_dist(hormone,sex,dtype='quantile'):
    file    = '../data/Clalit_seasonality_Nov19/clock.' + hormone +'.healthy_no_meds.' + dtype + '.with_infection.0.05.20_80.csv'
    df  = pd.read_csv(file)  
    df  = df.loc[df['type'] ==sex]
    
    df['hour'] = df['hour']+12
    df.loc[df['hour']>24,['hour']] = df.loc[df['hour']>24,['hour']] - 24
    df = df.loc[df['n']>20]
    
    wh= []
    for seas in ['winter','spring','summer','fall']:
        ddf = df.loc[df['season'] == seas].copy()
        ddf = ddf.sort_values('hour')
        wh.append(ddf['hour'].values)
    common_hours = list(set(wh[0]) & set(wh[1]) & set(wh[2])& set(wh[3]))    
    
    # get winter test time disturbution
    ddf = df.loc[df['season'] == 'winter'].copy()
    ddf = ddf.sort_values('hour')
    ddf = ddf.set_index('hour')
    ddf = ddf.loc[common_hours]
    P   = ddf['n']/ddf['n'].sum()
    
    seas_pooled_mean = []
    seas_pooled_std  = []
    seas_pooled_n    = []
    for seas in ['winter','spring','summer','fall']:
        ddf = df.loc[df['season'] == seas].copy()
        ddf = ddf.sort_values('hour')
        ddf = ddf.set_index('hour')
        ddf = ddf.loc[common_hours]
        ddf['p'] = P
        ntot = ddf['n'].sum()
        ddf = ddf[['mean','std','p']]; 
        
        seas_pooled_mean.append(np.sum(ddf['mean']*ddf['p']))
        seas_pooled_std.append(np.sqrt(np.sum(ddf['std']**2*(ddf['p']*ntot-1))/(np.sum(ddf['p']*ntot) - len(ddf['p']))))
        seas_pooled_n.append(ntot)

    both_res = pd.DataFrame({'seasons': ['winter','spring','summer','fall'],
                     'mean' : seas_pooled_mean,
                     'std'  : seas_pooled_std,
                     'n'    : seas_pooled_n})
    return both_res
```
