```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import glob
import calendar
import pickle
import os
from scipy.integrate import odeint
from matplotlib import ticker
from tqdm import tqdm_notebook as tqdm
from scipy import interpolate
from scipy.interpolate import interp1d
from scipy.stats import circmean,circstd
from scipy.optimize import curve_fit
from scipy.stats import zscore
from sympy import *
import matplotlib.image as mpimg
from  scipy.stats import distributions
from collections import namedtuple
from itertools import combinations
from matplotlib.backends.backend_pdf import PdfPages

def_color=plt.rcParams['axes.prop_cycle'].by_key()['color']
font = {'size'   : 10}

month_names=[x[:3] for x in calendar.month_name[1:13]]
plt.rc('font', weight='bold')

all_hormone_list=['ACTH_ADRENOCORTICOTROPIC_HORMONE',
                    'TSH_THYROID_STIMULATING_HORMONE',
                    'LH_LUTEINIZING_HORMONE',
                    'FSH_FOLLICLE_STIMULATING_HORMONE',
                    'GROWTH_HORMONE_GH',
                    'PROLACTIN',
                    'CORTISOL_BLOOD',
                    'CORTISOL_U_FREE_24h',
                    'T3_FREE',
                    'T4_FREE',
                    'ESTRADIOL_E_2',
                    'TESTOSTERONE_TOTAL',
                    'INSULIN_LIKE_GROWTH_FACTOR_1']

all_hormone_list_short=['ACTH',
                    'TSH',
                    'LH',
                    'FSH',
                    'GH',
                    'PRL',
                    'CORT',
                    'CORT24h',
                    'T3',
                    'T4',
                    'E2',
                    'TEST',
                    'IGF-1']
```

```python
pdf_pages = PdfPages('../figures/Seasonality24112019/aging.pdf')
for h,hormone in tqdm(enumerate(all_hormone_list)):
    
    fig = plt.figure(figsize=(10,3))
    
    for g,gender  in enumerate(['female','male']):

        file    = '../data/Clalit_seasonality_Nov19/age.' + hormone +'.healthy_no_meds.' + 'raw' + '.with_infection.0.05.20_30_40_50_60_70_80.csv'
        df = pd.read_csv(file)
        df = df.loc[df['type']==gender]

        pooled_median = []
        pooled_mean   = []
        pooled_std    = []
        pooled_n    = []

        for ag in df.age_group.unique():
            adf = df.loc[df['age_group']==ag]
            pooled_median.append(np.sum(adf['median']*adf['n'])/np.sum(adf['n']))
            pooled_mean  .append(np.sum(adf['mean']*adf['n'])/np.sum(adf['n']))
            pooled_std   .append(np.sqrt(np.sum(adf['std']**2*(adf['n']-1))/(np.sum(adf['n']) - len(adf) + 1)))       
            pooled_n     .append(np.sum(adf['n']))
        pooled_err = pooled_std/np.sqrt(pooled_n)   

        ax = plt.subplot(1,3,g+1)
        
        ax.errorbar(np.arange(len(pooled_mean)),pooled_mean,pooled_err,fmt='-o',c=def_color[0])   
        ax.set_ylabel('raw mean',fontweight='bold',c=def_color[0])

        axn = ax.twinx()
        
        axn.errorbar(np.arange(len(pooled_mean)),pooled_median,pooled_err,fmt='-o',c=def_color[1])   
        axn.set_ylabel('raw median',fontweight='bold',c=def_color[1])

        ax.set_xticks(np.arange(len(pooled_mean)))
        ax.set_xticklabels(df.age_group.unique(),rotation=45)
        ax.set_title(all_hormone_list_short[h] + ', ' + gender + ', N = ' +str(df.n.values.sum()),fontsize=12,fontweight='bold')
        
        ax = plt.subplot(1,3,3)
        ax.errorbar(np.arange(len(pooled_mean)),pooled_median,pooled_err,fmt='-o',c=def_color[g+4])   
        ax.set_xticks(np.arange(len(pooled_mean)))
        ax.set_xticklabels(df.age_group.unique(),rotation=45)
        ax.set_title('male-female median',fontweight='bold')
        ax.legend(['female','male'])
#         file    = '../data/Clalit_seasonality_Nov19/age.' + hormone +'.healthy_no_meds.' + 'quantile' + '.with_infection.0.05.20_30_40_50_60_70_80.csv'
#         df = pd.read_csv(file)
#         df = df.loc[df['type']==gender]
#         df = df.loc[df['age_group']!='[70,80]']
#         ax = plt.subplot(1,2,2)
#         season_mean_std = []
#         for j,ag in enumerate(df.age_group.unique()):
#             sdf = df.loc[df['age_group'] == ag]
#             hsdf= sdf.loc[sdf['type']==gender]
#             hsdf = hsdf.sort_values('month')
#             ax.errorbar(hsdf.month.values,hsdf['mean'].values,yerr = hsdf['std'].values/np.sqrt(hsdf.n.values),fmt='-o',c=def_color[j])   

#         ax.set_xlim([1,12])
#         ax.set_xticks(np.arange(1,13))
#         ax.set_xticklabels(month_names,rotation=90)
#         ax.set_ylabel('percentile [%]',fontweight='bold')
#         plt.legend(df.age_group.unique(),bbox_to_anchor=(1, 1))
    plt.tight_layout()
    pdf_pages.savefig(fig)

pdf_pages.close()
```
