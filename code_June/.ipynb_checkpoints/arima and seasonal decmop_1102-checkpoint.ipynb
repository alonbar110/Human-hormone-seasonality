```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import glob
import calendar
import pickle
import os
from scipy.integrate import odeint
from matplotlib import ticker
from tqdm import tqdm_notebook as tqdm
from scipy import interpolate
from scipy.interpolate import interp1d
from scipy.stats import circmean,circstd
from scipy.optimize import curve_fit
from scipy.stats import zscore
from sympy import *
from datetime import datetime
import pickle
from statsmodels.tsa.statespace.sarimax import SARIMAX
from datetime import datetime
from pyramid.arima import auto_arima
from statsmodels.tsa.seasonal import seasonal_decompose

effector_hormone_dic={'CORTISOL_BLOOD':'Cortisol',
                     'CORTISOL_U_FREE_24h':'Urinary cortisol',
                     'T3_FREE':'T3-free',
                     'T4_FREE':'T4-free',
                     'ESTRADIOL_E_2':'Estradiol',
                     'TESTOSTERONE_TOTAL':'Testosterone',
                     'INSULIN_LIKE_GROWTH_FACTOR_1':'IGF1'}

pituitary_hormone_dic={'ACTH_ADRENOCORTICOTROPIC_HORMONE':'ACTH',
                    'TSH_THYROID_STIMULATING_HORMONE':'TSH',
                    'LH_LUTEINIZING_HORMONE':'LH',
                    'FSH_FOLLICLE_STIMULATING_HORMONE':'FSH',
                    'GROWTH_HORMONE_GH':'GH',
                    'PROLACTIN':'Prolactin'}

all_hormone_list=['ACTH_ADRENOCORTICOTROPIC_HORMONE',
                    'TSH_THYROID_STIMULATING_HORMONE',
                    'LH_LUTEINIZING_HORMONE',
                    'FSH_FOLLICLE_STIMULATING_HORMONE',
                    'GROWTH_HORMONE_GH',
                    'PROLACTIN',
                    'CORTISOL_BLOOD',
                    'CORTISOL_U_FREE_24h',
                    'T3_FREE',
                    'T4_FREE',
                    'ESTRADIOL_E_2',
                    'TESTOSTERONE_TOTAL',
                    'INSULIN_LIKE_GROWTH_FACTOR_1']

def_color=plt.rcParams['axes.prop_cycle'].by_key()['color']

def get_longitudonal(hormone,sex,dtype='quantile'):
    file = '../data/Clalit_seasonality_Nov19/year.' + hormone +'.healthy_no_meds.' + dtype + '.with_infection.0.05.20_80.csv'
    df   = pd.read_csv(file)
    df   = df.loc[df['type']==sex]
    
    if sum(df['year'] == df['year'].min()) < 12:
        df = df.loc[df['year']>df['year'].min()]

    if sum(df['year'] == df['year'].max()) < 12:
        df = df.loc[df['year']<df['year'].max()]

    df['date'] = [datetime(df['year'].values[i],df['month'].values[i],1) for i in range(len(df))]
    df = df.sort_values('date')
    df = df.set_index(df['date'])
    return df

def data2xy(data):
    '''
    Take a seasonality data from a file and change the months to angle (0,2*pi)
    return the angle month and the mean for each month
    '''
    xdat=np.arange(1,13)
    ydat=np.array(data['mean'])#hormone value
    edat= np.array(data['std']/np.sqrt(data['n']))
    return xdat,ydat,edat

def add_months(sourcedate, months):
    month = sourcedate.month - 1 + months
    year = sourcedate.year + month // 12
    month = month % 12 + 1
    day = min(sourcedate.day, calendar.monthrange(year,month)[1])
    return datetime(year, month, day).date()

def AUTO_SARIMA_hormone(hormone,sex):
    df     = get_longitudonal(hormone,sex)
    data   = df['mean']
    model  = auto_arima(data,m=12,seasonal=True)
    return model

def seasonal_decompose_hormone(hormone,sex):
    df = get_longitudonal(hormone,sex)
    df = df.fillna(df.mean())
    xdat,ydat,edat = data2xy(df)

    boots = []
    for i in range(100):
        yboot = [np.random.normal(loc = y,scale = e) for y,e in zip(ydat,edat)]
        res = seasonal_decompose(yboot,freq=12)
        ses_boot = res.seasonal[:12]
        boots.append(ses_boot)
        
    boots    = np.array(boots)
    ses_mean = np.mean(boots,axis=0)
    ses_err  = np.std(boots,axis=0)
    plt.errorbar(xdat,ses_mean,ses_err,c=def_color[0])
#     plt.plot(xdat, ses_mean,c='r')
    plt.fill_between(xdat, ses_mean-ses_err, ses_mean+ses_err, alpha=0.35)
    
    month_pooled_mean = []
    month_pooled_std  = []
    month_pooled_n    = []

    for month in np.arange(1,13):
        mdf = df.loc[df.month==month]
        month_pooled_mean.append(np.sum(mdf['mean']*mdf['n'])/np.sum(mdf['n']))
        month_pooled_std.append(np.sqrt(np.sum(mdf['std']**2*(mdf['n']-1))/(np.sum(mdf['n']) - len(mdf))))
        month_pooled_n.append(np.sum(mdf['n']))

    fem_res = pd.DataFrame({'month': np.arange(0,12),
                             'mean' : month_pooled_mean,
                             'std'  : month_pooled_std,
                             'n'    : month_pooled_n})

#     plt.plot(xdat,np.array(month_pooled_mean) - np.array(month_pooled_mean).mean(),'--k')
#     plt.legend(['averaging','decmop'])
    plt.title(hormone)
    plt.show()
```

```python
[pituitary_list,effector_list] = pickle.load(open('seasonality_processed_data1102_20_80.pickle','rb'))[0:2]
```

```python
res = []
for hormone in tqdm(all_hormone_list):
    for sex in ['female','male']:
        model = AUTO_SARIMA_hormone(hormone,sex)
        res.append({'hormone':hormone, 'sex' : sex,'seasonal': sum(model.seasonal_order[:3]) > 0})
```

```python
arima_df = pd.DataFrame(res)
```
